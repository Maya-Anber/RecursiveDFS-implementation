# Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = {
    'A': ['B', 'C','D'],
    'B': ['E','F'],
    'C': [],
    'D': ['H','G'],
    'E': ['J','K'],
    'F': [],
    'G': [],
    'H': ['L','I'],
    'I': [],
    'J': [],
    'K': [],
    'L': [],
}
```

```python
G = nx.Graph()
for node, neighbors in graph.items():
    for neighbor in neighbors:
        G.add_edge(node, neighbor)
# Position the nodes using a layout algorithm
pos = nx.spring_layout(G)
# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_weight='bold')
plt.title('Graph')
plt.show()
```

# Recursive DFS implementation (traversing the graph)

Perform a recursive Depth First Search (DFS) on the graph(keeping track of the explored node)

>   graph: A dictionary representing the adjacency list of the graph

>   start: The starting node for the DFS

>   visited: A set of already visited nodes
    

```python
def dfs_recursive_traverse(graph, start):
    visited = []
    stack = [[start]]
    while stack:
        path = stack.pop()
        node = path[-1]
        if node in visited:
            continue
        visited.append(node)
        print(node)
        adjacent_nodes = graph.get(node, [])
        for node2 in adjacent_nodes:
            new_path = path.copy()
            new_path.append(node2)
            stack.append(new_path)
    return None
```

```python
print('visited nodes after the DFS traverse')
visited_nodes = dfs_recursive_traverse(graph, 'A')
```

# Performing DFS to reach a goal node

```python
def dfs_recursive_goal(graph, start, goal):
    visited = []
    stack = [[start]]
    while stack:
        path = stack.pop()
        node = path[-1]
        if node in visited:
            continue
        visited.append(node)
        print('Explored',node)
        if node == goal:
            print(f"Goal {goal} found!")
            return path
        else:
            adjacent_nodes = graph.get(node, [])
            for node2 in adjacent_nodes:
                new_path = path.copy()
                new_path.append(node2)
                stack.append(new_path)
    print(f"Node {goal} not found!")
    return None
```

```python
solution=dfs_recursive_goal(graph, 'A', 'E')     
print("The path solution is",solution)
```

# VISUALIZATION

```python
#GRAPH VISUALIZATION
def dfs_graph_goal(graph, start, goal):
    visited = []
    stack = [[start]]
    while stack:
        path = stack.pop()
        node = path[-1]
        if node in visited:
            continue
        visited.append(node)
        if node == goal:
            return visited, path
        else:
            adjacent_nodes = graph.get(node, [])
            for node2 in adjacent_nodes:
                new_path = path.copy()
                new_path.append(node2)
                stack.append(new_path)
    return visited, None
```

```python
# Define the start and goal nodes
start_node = 'A'
goal_node = 'E'

# Perform DFS and capture the traversal path and visited nodes
visited_nodes, dfs_path = dfs_graph_goal(graph, start_node, goal_node)
# Draw the graph
plt.figure(figsize=(10, 8))
# Draw nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_weight='bold')

# Highlight the visited nodes
nx.draw_networkx_nodes(G, pos, nodelist=visited_nodes, node_color='r', node_size=2000)

# Highlight the DFS traversal path
if dfs_path:
    edges = [(dfs_path[i], dfs_path[i + 1]) for i in range(len(dfs_path) - 1)]
    nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color='gold', width=2)

# Save the visualization as an image file
plt.title('DFS Traversal with Goal Node Highlighted')
plt.show()
```
